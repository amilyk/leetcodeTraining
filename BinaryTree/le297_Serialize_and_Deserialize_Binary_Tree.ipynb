```python
###层次遍历 及 反序列化######################
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Codec:

    def serialize(self, root):
        """Encodes a tree to a single string.
        
        :type root: TreeNode
        :rtype: str
        """
        #层次遍历
        queue = [root]
        res = []

        if not root:
            return 'x'
        while queue:
            cur = queue.pop(0)
            if cur:
                res.append(str(cur.val))
            else:
                res.append('x')
                continue #如果cur为None，则cur.left cur.right不存在，不做append操作
            queue.append(cur.left)
            queue.append(cur.right)
        return ",".join(res)

    
    def deserialize(self, data):
        """Decodes your encoded data to tree.
        
        :type data: str
        :rtype: TreeNode
        """
        #层级遍历结果反序列化二叉树
        #关键在于每次都能找到parent
        if data[0] == 'x':
            return None
        queue = []
        datalist = data.split(',')
        root = TreeNode(datalist.pop(0))
        queue.append(root)

        while len(datalist) > 0:
            parent = queue.pop(0)
            left_val = datalist.pop(0)
            if left_val != 'x':
                parent.left = TreeNode(left_val)
                queue.append(parent.left)
            else:
                parent.left = None

            right_val = datalist.pop(0)
            if right_val != 'x':
                parent.right = TreeNode(right_val)
                queue.append(parent.right)
            else:
                parent.right = None

        return root
    
        

# Your Codec object will be instantiated and called as such:
# ser = Codec()
# deser = Codec()
# ans = deser.deserialize(ser.serialize(root))
```
